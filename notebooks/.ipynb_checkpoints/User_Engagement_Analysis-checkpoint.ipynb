### Normalizing each engagement metric and run a k-means (k=3) to classify customers in three groups of engagement

```python
km = KMeans(n_clusters=3,)
```

```python
#Normalizing the dataframe
engagement_df_norm = (engagement_df-engagement_df.min())/ (engagement_df.max() - engagement_df.min())
```

```python
engagement_df_norm
```

```python
#Determining the clusters to which the datapoints are containedb
prediction = km.fit_predict(engagement_df_norm[['Total_Data', 'Dur_s', 'Session frequency']])
prediction
```

```python
#Creating a new variable to display various clusters
engagement_df_norm['cluster'] = prediction 

engagement_df_norm1 = engagement_df_norm[engagement_df_norm.cluster==0]
engagement_df_norm2 = engagement_df_norm[engagement_df_norm.cluster==1]
engagement_df_norm3 = engagement_df_norm[engagement_df_norm.cluster==2]

engagement_df_norm
```

```python
#Computing metric for each required variable
variables = ['Total_Data', 'Dur_s', 'Session frequency']


print('CLUSTER 0')
for column in variables:
    print(engagement_df_norm1[column].describe())
print('CLUSTER 1')
for column in variables:
    print(engagement_df_norm2[column].describe())
print('CLUSTER 2')
for column in variables:
    print(engagement_df_norm3[column].describe())
```

Aggregate user total traffic per application and derive the top 10 most engaged users per application

```python
engagement_df_norm['Google_total(MB)'] = (data['Google_DL_(Bytes)'] + data['Google_UL_(Bytes)'])/10**6
engagement_df_norm['Social_total(MB)'] = (data['Social_Media_DL_(Bytes)'] + data['Social_Media_UL_(Bytes)'])/10**6
engagement_df_norm['Email_total(MB)'] = (data['Email_DL(Bytes)'] + data['Email_UL(Bytes)'])/10**6
engagement_df_norm['Youtube_total(MB)'] = (data['Youtube_DL_(Bytes)'] + data['Youtube_UL_(Bytes)'])/10**6
engagement_df_norm['Netflix_total(MB)'] = (data['Netflix_DL_(Bytes)'] + data['Netflix_UL_(Bytes)'])/10**6
engagement_df_norm['Gaming_total(MB)'] = (data['Gaming_DL_(Bytes)'] + data['Gaming_UL_(Bytes)'])/10**6
engagement_df_norm['Other_total(MB)'] = (data['Other_DL_(Bytes)'] + data['Other_UL_(Bytes)'])/10**6
```

```python
engagement_df_norm.info()
```

```python
relevant_apps = ['Google_total(MB)','Social_total(MB)', 'Email_total(MB)', 'Youtube_total(MB)','Netflix_total(MB)', 'Gaming_total(MB)']
```

```python

traffic_data = engagement_df_norm.groupby('MSISDN_Number')
for app in relevant_apps:
    traffic_data.get_group(app).value_counts()[0:10]
```

```python
top_3 = ['Gaming_total(MB)', 'Youtube_total(MB)', 'Google_total(MB)']
for cols in top_3:
    sns.histplot(data=engagement_df_norm, x=cols)
    plt.show()
```

```python
#Creating a new dataframe to work on the k-means clustering algorithm

dfs = pd.DataFrame()
dfs['Total_Data'] = engagement_df_norm['Total_Data']
dfs['Session frequency'] = engagement_df_norm['Session frequency']
dfs['Dur_s'] = engagement_df_norm['Dur_s']
```

```python
from scipy.spatial.distance import cdist
```

```python
#Creating and fitting the model for eucledean distance calculation
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 10)
for k in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(dfs)
    kmeanModel.fit(dfs)
 
    distortions.append(sum(np.min(cdist(dfs, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / dfs.shape[0])
    inertias.append(kmeanModel.inertia_)
 
    mapping1[k] = sum(np.min(cdist(dfs, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / dfs.shape[0]
    mapping2[k] = kmeanModel.inertia_
```

```python
plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

### Interpetation

The optimized value of K is 3 hence the data is best classified into those 3 clusters
