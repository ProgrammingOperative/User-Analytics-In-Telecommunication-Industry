```python
'''Aggregate per user using the selected columns as below - Including creating of new rows and 
changing the the display of the data to float of one decimal place
'''


pd.set_option('display.float_format', lambda x: '%.1f' % x)

clean_data['Total_UL_and_DL_(Bytes)'] = clean_data['Total_UL_(Bytes)'] + clean_data['Total_DL_(Bytes)']
clean_data['Other_UL_and_DL_(Bytes)'] = clean_data['Other_UL_(Bytes)'] + clean_data['Other_DL_(Bytes)']
clean_data['Gaming_UL_and_DL_(Bytes)'] = clean_data['Gaming_UL_(Bytes)'] + clean_data['Gaming_DL_(Bytes)']
clean_data['Netflix_UL_and_DL_(Bytes)'] = clean_data['Netflix_UL_(Bytes)'] + clean_data['Netflix_DL_(Bytes)']
clean_data['Youtube_UL_and_DL_(Bytes)'] = clean_data['Youtube_UL_(Bytes)'] + clean_data['Youtube_DL_(Bytes)']
clean_data['Email_UL_and_DL_(Bytes)'] = clean_data['Email_UL(Bytes)'] + clean_data['Email_DL(Bytes)']
clean_data['Google_UL_and_DL_(Bytes)'] = clean_data['Google_UL_(Bytes)'] + clean_data['Google_DL_(Bytes)']
clean_data['Social Media_UL_and_DL_(Bytes)'] = clean_data['Social_Media_UL_(Bytes)'] + clean_data['Social_Media_DL_(Bytes)']
clean_data.head()
```

```python
#Selecting a few of the columns for overview analysis of the user behaviour

user_columns = ['Bearer_Id','Dur_ms','MSISDN_Number','Total_UL_and_DL_(Bytes)','Other_UL_and_DL_(Bytes)',
                'Gaming_UL_and_DL_(Bytes)','Netflix_UL_and_DL_(Bytes)','Youtube_UL_and_DL_(Bytes)',
                'Email_UL_and_DL_(Bytes)','Google_UL_and_DL_(Bytes)','Social Media_UL_and_DL_(Bytes)']

user_overview_data = clean_data[user_columns]
user_overview_data.head()
```

```python
# Converting the data capacity from bytes to Megabytes
for column in user_columns:
    if column not in ['Bearer_Id','Dur_ms','MSISDN_Number']:
        user_overview_data[column]= user_overview_data[column]/(10**6)

user_overview_data.head()
```

```python
#Aggregating all column data about every user
users = user_overview_data.groupby(['MSISDN_Number'])#First group the data by user
users_aggregated = users.agg({'Bearer_Id':'count','Dur_ms':'sum','Total_UL_and_DL_(Bytes)':'sum',
                              'Other_UL_and_DL_(Bytes)':'sum','Gaming_UL_and_DL_(Bytes)':'sum',
                              'Netflix_UL_and_DL_(Bytes)':'sum','Youtube_UL_and_DL_(Bytes)':'sum',
                              'Email_UL_and_DL_(Bytes)':'sum','Google_UL_and_DL_(Bytes)':'sum',
                              'Social Media_UL_and_DL_(Bytes)':'sum'})
users_aggregated
```

```python
#Fixing identified outlier
def fix_outlier(df, column):
    df[column] = np.where(df[column] > df[column].quantile(0.95), df[column].median(),df[column])
    
    return df[column]


user_overview_data['Dur_ms']=fix_outlier(user_overview_data,'Dur_ms')
user_overview_data['Total_UL_and_DL_(Bytes)']=fix_outlier(user_overview_data,'Total_UL_and_DL_(Bytes)')
user_overview_data['Other_UL_and_DL_(Bytes)']=fix_outlier(user_overview_data,'Other_UL_and_DL_(Bytes)')
user_overview_data['Gaming_UL_and_DL_(Bytes)']=fix_outlier(user_overview_data,'Gaming_UL_and_DL_(Bytes)')
user_overview_data['Netflix_UL_and_DL_(Bytes)']=fix_outlier(user_overview_data,'Netflix_UL_and_DL_(Bytes)')
user_overview_data['Youtube_UL_and_DL_(Bytes)']=fix_outlier(user_overview_data,'Youtube_UL_and_DL_(Bytes)')
user_overview_data['Email_UL_and_DL_(Bytes)']=fix_outlier(user_overview_data,'Email_UL_and_DL_(Bytes)')
user_overview_data['Google_UL_and_DL_(Bytes)']=fix_outlier(user_overview_data,'Google_UL_and_DL_(Bytes)')
user_overview_data['Social_Media_UL_and_DL_(Bytes)']=fix_outlier(user_overview_data,'Social_Media_UL_and_DL_(Bytes)')
```

```python
###################################PLOTTING FUNCTIONS###################################

def plot_hist(df:pd.DataFrame, column:str, color:str)->None:
    plt.figure(figsize=(15, 10))
    fig, ax = plt.subplots(1, figsize=(12, 7))
    sns.displot(data=df, x=column, color=color, kde=True, height=7, aspect=2)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.show()

def plot_count(df:pd.DataFrame, column:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.show()
    
def plot_bar(df:pd.DataFrame, x_col:str, y_col:str, title:str, xlabel:str, ylabel:str)->None:
    plt.figure(figsize=(12, 7))
    sns.barplot(data = df, x=x_col, y=y_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.yticks( fontsize=14)
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.show()

def plot_heatmap(df:pd.DataFrame, title:str, cbar=False)->None:
    plt.figure(figsize=(12, 7))
    sns.heatmap(df, annot=True, cmap='viridis', vmin=0, vmax=1, fmt='.2f', linewidths=.7, cbar=cbar )
    plt.title(title, size=18, fontweight='bold')
    plt.show()

def plot_box(df:pd.DataFrame, x_col:str, title:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.boxplot(data = df, x=x_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.show()

def plot_box_multi(df:pd.DataFrame, x_col:str, y_col:str, title:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.boxplot(data = df, x=x_col, y=y_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.yticks( fontsize=14)
    plt.show()

def plot_scatter(df: pd.DataFrame, x_col: str, y_col: str, title: str) -> None:
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data = df, x=x_col, y=y_col)
    plt.title(title, size=20)
    plt.xticks(fontsize=14)
    plt.yticks( fontsize=14)
    plt.show()
    
def plot_univariate(df, col1, col2):
    plt.figure(figsize = (20, 15))

    plt.subplot(2, 2, 1)
    plt.hist(df[col1], bins = 20, color = 'green', edgecolor = 'gray', linewidth = 0.5)
    plt.title(f'Histogram of {col1}', size=14)

    plt.subplot(2, 2, 2)
    plt.boxplot(df[col1])
    plt.title(f'Boxplot of {col1}', size=14)

    plt.subplot(2, 2, 3)
    plt.hist(df[col2], bins=20, color='green', edgecolor='gray', linewidth=0.5)
    plt.title(f'Histogram of {col2}', size=14)

    plt.subplot(2, 2, 4)
    plt.boxplot(df[col2])
    plt.title(f'Boxplot of {col2}', size=14)
    
       

    plt.show()
```

```python
#Plotting the univariate Graphical Interpretation

plot_univariate(user_overview_data,'Dur_ms','Total_UL_and_DL_(Bytes)')
plot_univariate(user_overview_data,'Other_UL_and_DL_(Bytes)','Gaming_UL_and_DL_(Bytes)')
plot_univariate(user_overview_data,'Netflix_UL_and_DL_(Bytes)','Youtube_UL_and_DL_(Bytes)')
plot_univariate(user_overview_data,'Email_UL_and_DL_(Bytes)','Google_UL_and_DL_(Bytes)')
plot_univariate(user_overview_data,'Gaming_UL_and_DL_(Bytes)')
```
